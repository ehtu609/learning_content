```python
agg_obj_1 = filtered_df.groupby(['just_date', 'Subject'],as_index=False)
agg_df_1 = pd.DataFrame(agg_obj_1.agg({'Email Address' : 'nunique'}))
agg_df_1.rename(columns={'Email Address':'Unique Recipients'}, inplace=True)
print(agg_df_1.shape)
print(agg_df_1)
print()
agg_obj_2 = filtered_df.groupby(['just_date', 'Subject'],as_index=False)
agg_df_2 = pd.DataFrame(agg_obj_2.agg({'Email Address' : 'count'}))
agg_df_2.rename(columns={'Email Address':'Total Recipients'}, inplace=True)
print(agg_df_2.shape)
print(agg_df_2.head())
final_df = pd.merge(agg_df_1, agg_df_2, left_on=['just_date', 'Subject'], right_on=['just_date', 'Subject'], how='inner')
final_df
```
